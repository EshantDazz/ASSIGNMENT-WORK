### 10. How can you tell the difference between break and continue?### 10. How can you tell the difference between break and continue?

### Answer:

If we use the break statement it will basically terminate the loop and no furthur iteration will take place but in case of continue the loop will skip the next following lines and will keep contiuing its iteration<br>
**Example**

```python
#break
for i in range(4):
    if i==2:
        break
    print(i)
```

It has printed only till 1 becasue when it saw the value of 2 the break operation took place and terminated

```python
#continue
for i in range(4):
    if i==2:
        continue
    print(i)
```

The termination was not stopped but when the value of i was 2 it skipped the next printing line and the value of i increamented and the iteration continued and we got our answer 0,1,3
